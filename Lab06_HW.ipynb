```python
#1
so = input("Nhập các số nguyên:").split()
unique = []
for value in so:
  if value in unique:
    continue
  else:
    unique.append(value)
print(list(map(int, unique)))
```

```python
#2
def sum(n: list[int]):
  sum = 0
  for i in n:
    sum += i
  return sum

so = input("Nhập các số nguyên:").split()
output = []
for i in range(0,len(so)):
  output.append(sum(list(map(int, so[0:i+1]))))
print(output)
```

```python
#3
so = tuple(input("Nhập các số nguyên:").split())
k = int(input("Nhập k:"))
new = []
for i in range(0,len(so)):
  w = (i+k)%len(so)
  new.append(so[w])
print(tuple(map(int,new)))
```

```python
#5
so = map(int,input("Nhập một list số nguyên: ").split())
pos = 0
neg = 0
zero = 0
for i in so:
  if i > 0: pos += 1
  if i < 0: neg += 1
  if i == 0: zero += 1
dic = {"positive":pos,"negative":neg,"zero":zero}
print(dic)
```

```python
#9
x = input("Nhập vào dictionary A: ").split()
y = input("Nhập vào dictionary B: ").split()
z = {}
for i in x:
  a,b = i.split(":")
  if a not in z:
    z[a] = int(b)
  else:
    z[a] += int(b)

for i in y:
  a,b = i.split(":")
  if a not in z:
    z[a] = int(b)
  else:
    z[a] += int(b)
print(z)
```

```python
#10
so = list(map(int,input("Nhập một list số nguyên: ").split()))
k = int(input("Nhập vào một số k: "))
uniq = set()
noloop = []
for a in so:
  i = so.index(a)
  sbr = so[0:i] + [0] + so[i+1:]
  if a not in noloop:
    noloop.append(a)
    for b in sbr:
      if a+b == k:
        tup = (a,b)
        uniq.add(tuple(sorted(tup)))

print(list(sorted(uniq)))
```

```python
#12
so = list(map(int,(input("Nhập các số nguyên: ")).split()))
uniq = {}

#Tạo dict với keys = số, values = số lần xuất hiện
for i in so:
  if i not in uniq:
    uniq[i] = 1
  else:
    uniq[i] += 1

#Tìm max value (số lần xuất hiện lớn nhất)
max = 0
for a in uniq.values():
  if a > max:
    max = a
maxvalue = [k for k in uniq.keys() if uniq[k] == max]
print(min(maxvalue))

```

```python
#19
def max(n: list[int]):
  maxx = 0
  for i in n:
    if i > maxx:
      maxx = i
  return maxx

so = list(map(int,(input("Nhập các số nguyên: ")).split()))
print(max(so))
```

```python
#20
def find(n: list[int], k: int):
  for i in n:
    if i == k:
      return n.index(i)+1
  return -1

so = list(map(int,(input("Nhập các số nguyên: ")).split()))
k = int(input("Nhập một số nguyên: "))
print(find(so,k))

```
