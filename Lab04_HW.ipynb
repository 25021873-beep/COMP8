```python
#6
num = input("Nhập hai số a, b: ")
a, b = num.split()
snt = []

#Thử mọi x thuộc [a,b] có phải số nguyên tố không
for i in range(int(a), int(b)+1):
  for k in range(2, int(i)):
      prime = True
      if i % k == 0:
        prime = False
        break

#Nếu có thì thêm vào list số nguyên tố
  if prime == True:
      snt.append(i)

#Tính tổng các thành phần trong list số nguyên tố
total = 0
for value in snt:
  total += value
print(total)


```

```python
#7
num = input("Nhập một số bất kỳ: ")

def prime(num):
  prime = True
  for i in range(2, int(num)):
    if int(num) % i == 0:
      prime = False
      break
  return prime

def uoc(x):
  uocc = []
  for i in range(2, int(x)+1):
    if int(x) % i == 0:
      uocc.append(i)

  for value in uocc:
    uocc = reversed(uocc)
    for value in uocc:
      if prime(value) == True:
        return value


if prime(num) == True:
  print(num)
else:
  print(uoc(num))
```

```python
#32
text = input("Nhập một cái gì đó: ")
has_numeric = False
has_upper = False
has_lower = False
has_special = False

if len(text) > 6:
    for i in text:
        if i.isnumeric():
            has_numeric = True
        elif i.isupper():
            has_upper = True
        elif i.islower():
            has_lower = True
        elif not i.isalnum() and i != " ":
            has_special = True

    if has_numeric and has_upper and has_lower and has_special:
        print("Mạnh")
    else:
        print("Yếu")
else:
    print("Yếu")
```
