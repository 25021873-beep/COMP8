```python
#3
def IsPrime(n):
  for i in range(2,int(n)):
    if n % i == 0:
      return False
  return True

n = int(input("Nhập số: "))
IsPrime(n)
```

```python
#4
def PerfectNum(n):
  uoc = []
  value = 0
  for i in range(1,int(n)):
    if n % i == 0:
      uoc.append(i)
  for k in uoc:
    value += k
  if value == n:
    return True
  else:
    return False

PerfectNum(24)
```

```python
#5
def ListIndex(dayso,n):
  for i in dayso:
    if i == n:
      return dayso.index(i)+1
  return -1


ListIndex([1,2,3,4,5,67,76,8],-6)
```

```python
#6
def Factorial(n):
  tich = 1
  for i in range(1,n+1):
    tich = tich * i
  return tich

Factorial(10)
```

```python
#7
def Calculator(a):
  for i in a:
    if i == "+":
      parts = a.split('+')
      return float(parts[0]) + float(parts[1])
    if i == "-":
      parts = a.split('-')
      return float(parts[0]) - float(parts[1])
    if i == "*":
      parts = a.split('*')
      return round((float(parts[0]) * float(parts[1])),2)
    if i == "/":
      parts = a.split('/')
      return round((float(parts[0]) / float(parts[1])),2)
  return "Invalid expression"



Calculator("12*-23.004")
```

```python
#8
def Hamming(a,b):
  count = 0
  # Đổi từ số thành chuỗi nhị phân, bỏ "0b" ở đầu:
  bin_a = bin(a)[2:]
  bin_b = bin(b)[2:]

  # Độn thêm 0 vào đầu chuỗi bé hơn để độ dài hai chuỗi bằng nhau:
  max_len = max(len(bin_a), len(bin_b))
  bin_a = bin_a.zfill(max_len)
  bin_b = bin_b.zfill(max_len)

  # Dùng for check so sánh từng giá trị trên mỗi chuỗi, nếu khác nhau thì count += 1:
  for i in range(max_len):
    if bin_a[i] != bin_b[i]:
      count += 1
  return count
```

```python
#9
def sum(n):
  sum = 0
  for i in str(n):
    sum += int(i)
  return sum

sum(98765)
```
